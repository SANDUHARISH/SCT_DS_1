# Data Analysis on Indian Cities Dataset



```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV
df = pd.read_csv('cities_r2.csv')  # adjust path if needed

# View the data
df.head()
```

```python
print(df.columns)
```

## 1. Top 10 Cities by Total Population


```python
top_cities = df.sort_values('population_total', ascending=False).head(10)

import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.bar(top_cities['name_of_city'], top_cities['population_total'], color='orange')
plt.xticks(rotation=45)
plt.title('Top 10 Cities by Total Population')
plt.xlabel('City')
plt.ylabel('Total Population')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

## 2.Top 10 Cities by Gender-wise Literacy Rate


```python
top_cities = df.sort_values('effective_literacy_rate_total', ascending=False).head(10)

import numpy as np

x = np.arange(len(top_cities))
width = 0.3

plt.figure(figsize=(12,6))
plt.bar(x - width, top_cities['effective_literacy_rate_male'], width, label='Male Literacy Rate', color='dodgerblue')
plt.bar(x, top_cities['effective_literacy_rate_female'], width, label='Female Literacy Rate', color='hotpink')
plt.bar(x + width, top_cities['effective_literacy_rate_total'], width, label='Total Literacy Rate', color='mediumseagreen')

plt.xticks(x, top_cities['name_of_city'], rotation=45)
plt.ylabel('Literacy Rate (%)')
plt.title('Top 10 Cities by Literacy Rate')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

 ## 3.Top 10 cities by  Sex Ratio Comparison 

```python
top_sex_ratio = df.sort_values('sex_ratio', ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.bar(top_sex_ratio['name_of_city'], top_sex_ratio['sex_ratio'], color='plum')
plt.xticks(rotation=45)
plt.title('Top 10 Cities by Sex Ratio')
plt.xlabel('City')
plt.ylabel('Sex Ratio (Females per 1000 Males)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
